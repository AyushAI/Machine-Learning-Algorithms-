```python
import pandas as pd
```

```python
dataset = pd.read_csv("bangalore house price prediction OHE-data.csv")
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
# now split the datasets into "x" and "y" variable
```

```python
x = dataset.drop('price',axis=1)
y = dataset['price']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
krg = KNeighborsRegressor(n_neighbors=5)
```

```python
krg
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
krg.fit(x_train,y_train)
```

```python
print("Training set score : {}".format(krg.score(x_train,y_train)))
print("Test set score : {}".format(krg.score(x_test,y_test)))
```
